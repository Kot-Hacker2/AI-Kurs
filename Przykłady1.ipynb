```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt

rolls_count = []
dice_value = []

dice_rolls = np.array([random.randint(1,7) for _ in range(20)])
print(dice_rolls)

for i in set(dice_rolls):
    dice_value.append(i)
    rolls_count.append(int( len(np.where(dice_rolls == i)[0])  ))

plt.bar(dice_value, rolls_count)
plt.title("Histogram rzutw koci")
plt.xlabel("warto")
plt.ylabel("liczba wystpie")
plt.legend()
plt.show()
```
